```python
#box plot
import matplotlib.pyplot as plt
x=[10,20,30,40,50,70,130]
plt.boxplot(x,notch=True,vert=False,widths=0.3,label=['Group A'],patch_artist=True,showmeans=True,sym="g",boxprops={"color":"red"},whiskerprops={"color":"green"})
plt.legend()
```

```python
import matplotlib.pyplot as plt
#multi-box plot
x=[10,20,30,40,50,70,130]
x1=[15,25,35,45,55,80,1]
y=[x,x1]
plt.boxplot(y,notch=True,vert=True,widths=0.3,label=['Group A',"GroupB"],patch_artist=True,showmeans=True,sym="g",boxprops={"color":"red"},whiskerprops={"color":"green"})
plt.legend()
```
