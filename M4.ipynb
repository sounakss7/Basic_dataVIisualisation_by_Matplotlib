```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=["c","c++","java","python","ruby"]
plt.pie(x, labels=y, autopct='%1.1f%%',shadow=True,radius=0.9,textprops={"fontsize": 12},wedgeprops={"linewidth": 1.5, "edgecolor": "brown"})
plt.title("Programming Language Popularity")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
x1=[2,3,4,5,6]
y=["c","c++","java","python","ruby"]
plt.pie(x, labels=y, autopct='%1.1f%%',shadow=True,radius=1.1,textprops={"fontsize":10},wedgeprops={"linewidth": 1.5, "edgecolor": "brown"})
plt.pie(x1,radius=0.5,autopct='%0.1f%%',wedgeprops={"linewidth": 1, "edgecolor": "black"})
plt.title("Programming Language Popularity")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```
