```python
import matplotlib.pyplot as plt
import numpy as np
import random
#histogram
var=np.random.randint(10,60,(50))
print(var)



```

```python
var2=[12,13,56,78,90,34,5,7,68,4,3,24,2,44,58,9,7,98,89,80,81,83,82,85,86,87]
plt.title('Histogram')
plt.xlabel('Marks')
plt.ylabel('Frequency')


plt.hist(var2,color="red",edgecolor="black",bins=10)

plt.show()




```

```python
plt.hist(var2,"auto",(0,150),edgecolor="black",cumulative=-1)
```

```python
plt.hist(var2,"auto",(0,150),edgecolor="black",cumulative=-1,histtype="barstacked",orientation="vertical",rwidth=0.8,log=True)
plt.xlabel('Marks')
plt.legend()
plt.axvline(x=np.mean(var2), color='blue', label='Mean')
plt.grid()
```
