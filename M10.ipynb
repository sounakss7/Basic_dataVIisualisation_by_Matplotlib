```python
import matplotlib.pyplot as plt
x=["message","kaka","ronaldo","neymar","suarez"]
y=[2,4,6,8,10]
plt.subplot(2,2,1)
plt.plot(y,x,color="red",label="Line 1")
x1=[1,2,3,4,5]
plt.subplot(2,2,2)
plt.pie(x1, labels=x, autopct='%1.1f%%', shadow=True)
x2=["ascii","heart","diamond","spade","club"]
y2=[4,8,12,16,20]
plt.subplot(2,2,3)
plt.bar(x2,y2,color="blue",label="Bar 1")

```
