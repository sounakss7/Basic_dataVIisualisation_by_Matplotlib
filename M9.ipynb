```python
#Fill_between_plots(main)
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6,7,8]
y=[2,3,4,5,3,2,7,5]
plt.plot(x,y,label='python',color='green')
plt.title("python plot")
plt.fill_between(x=[2,4],y1=[3,5],y2=[4,2],color='lightblue',label='python area')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
```
